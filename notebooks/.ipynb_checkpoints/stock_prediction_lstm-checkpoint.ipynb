# 📈 삼성전자 주가 예측 - LSTM 딥러닝

이 노트북은 LSTM(Long Short-Term Memory) 신경망을 사용해서
삼성전자 주가의 다음 날 종가를 예측합니다.

**흐름:**
1. 데이터 수집 (FinanceDataReader)
2. 탐색적 데이터 분석 (EDA)
3. 데이터 전처리
4. LSTM 모델 구축 및 학습
5. 예측 결과 시각화 및 평가

```python
# ─────────────────────────────────────────
# 1단계: 필요한 라이브러리 불러오기
# ─────────────────────────────────────────
import FinanceDataReader as fdr   # 한국 주가 데이터 수집
import pandas as pd               # 데이터 처리
import numpy as np                # 수치 계산
import matplotlib.pyplot as plt   # 그래프 그리기
import matplotlib
import seaborn as sns             # 예쁜 그래프
import warnings

from sklearn.preprocessing import MinMaxScaler  # 데이터 정규화 (0~1 사이로 변환)
from sklearn.metrics import mean_squared_error, mean_absolute_error  # 성능 평가 지표

from tensorflow.keras.models import Sequential  # 순차 모델 (레이어를 순서대로 쌓는 방식)
from tensorflow.keras.layers import LSTM, Dense, Dropout  # 신경망 레이어
from tensorflow.keras.callbacks import EarlyStopping  # 과적합 방지: 성능 개선 없으면 자동 중단

# 한글 폰트 설정 (그래프에서 한글 깨짐 방지)
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
warnings.filterwarnings('ignore')

print('라이브러리 로드 완료!')
```

```python
# ─────────────────────────────────────────
# 2단계: 데이터 수집
# ─────────────────────────────────────────
# 삼성전자 종목코드: 005930
# FinanceDataReader로 한국거래소 데이터를 직접 가져옴

TICKER = '005930'         # 삼성전자 종목코드
START_DATE = '2019-01-01' # 데이터 시작일
END_DATE = '2024-12-31'   # 데이터 종료일

df = fdr.DataReader(TICKER, START_DATE, END_DATE)

print(f'데이터 수집 완료: {len(df)}개 거래일')
print(f'기간: {df.index[0].date()} ~ {df.index[-1].date()}')
print('\n첫 5행 미리보기:')
df.head()
```

```python
# ─────────────────────────────────────────
# 3단계: 탐색적 데이터 분석 (EDA)
# ─────────────────────────────────────────

# 기본 통계량 확인
print('=== 기본 통계량 ===')
print(df[['Open', 'High', 'Low', 'Close', 'Volume']].describe())

# 결측값 확인
print('\n=== 결측값 확인 ===')
print(df.isnull().sum())
```

```python
# 주가 흐름 시각화
fig, axes = plt.subplots(2, 1, figsize=(14, 8))

# 종가 추이
axes[0].plot(df.index, df['Close'], color='steelblue', linewidth=1)
axes[0].set_title('삼성전자 종가 추이 (2019~2024)', fontsize=14)
axes[0].set_ylabel('주가 (원)')
axes[0].grid(True, alpha=0.3)

# 거래량 추이
axes[1].bar(df.index, df['Volume'], color='salmon', alpha=0.6, width=1)
axes[1].set_title('일별 거래량', fontsize=14)
axes[1].set_ylabel('거래량 (주)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ─────────────────────────────────────────
# 4단계: 이동평균선 추가 (기술적 지표)
# ─────────────────────────────────────────
# 이동평균선: 최근 N일간 평균값 → 주가 추세 파악에 사용

df['MA_20'] = df['Close'].rolling(window=20).mean()   # 20일 이동평균
df['MA_60'] = df['Close'].rolling(window=60).mean()   # 60일 이동평균
df['MA_120'] = df['Close'].rolling(window=120).mean() # 120일 이동평균

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['Close'], label='종가', color='black', linewidth=1, alpha=0.7)
plt.plot(df.index, df['MA_20'], label='20일 이동평균', color='blue', linewidth=1.5)
plt.plot(df.index, df['MA_60'], label='60일 이동평균', color='orange', linewidth=1.5)
plt.plot(df.index, df['MA_120'], label='120일 이동평균', color='red', linewidth=1.5)
plt.title('삼성전자 주가 + 이동평균선', fontsize=14)
plt.ylabel('주가 (원)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# ─────────────────────────────────────────
# 5단계: 데이터 전처리
# ─────────────────────────────────────────

# 종가(Close)만 사용해서 예측
data = df[['Close']].copy()

# MinMaxScaler: 주가를 0~1 사이 값으로 변환
# 신경망은 큰 숫자에 민감하므로 정규화 필수
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# 훈련/테스트 데이터 분리 (80:20 비율)
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

print(f'전체 데이터: {len(scaled_data)}일')
print(f'훈련 데이터: {len(train_data)}일')
print(f'테스트 데이터: {len(test_data)}일')

# 슬라이딩 윈도우로 입력/출력 데이터 생성
# 예시: 과거 60일 데이터 → 다음 날 예측
LOOKBACK = 60  # 입력으로 사용할 과거 일수

def create_sequences(data, lookback):
    """슬라이딩 윈도우로 X(입력), y(정답) 쌍 생성"""
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, 0])  # 과거 60일
        y.append(data[i, 0])             # 다음 날 종가
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_data, LOOKBACK)
X_test, y_test = create_sequences(test_data, LOOKBACK)

# LSTM은 3D 입력 필요: (샘플 수, 타임스텝 수, 특성 수)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

print(f'\nX_train 형태: {X_train.shape}  → (샘플수, 60일, 1개 특성)')
print(f'X_test 형태: {X_test.shape}')
```

```python
# ─────────────────────────────────────────
# 6단계: LSTM 모델 구축
# ─────────────────────────────────────────

model = Sequential([
    # 첫 번째 LSTM 레이어: 128개 뉴런, return_sequences=True → 다음 LSTM에 전달
    LSTM(units=128, return_sequences=True, input_shape=(LOOKBACK, 1)),
    Dropout(0.2),  # 20% 뉴런을 무작위로 끄기 → 과적합 방지

    # 두 번째 LSTM 레이어: 64개 뉴런
    LSTM(units=64, return_sequences=False),
    Dropout(0.2),

    # 출력 레이어: 1개 값(다음 날 종가) 예측
    Dense(units=1)
])

# 모델 컴파일: 손실함수=평균제곱오차, 최적화=Adam
model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
# ─────────────────────────────────────────
# 7단계: 모델 학습
# ─────────────────────────────────────────

# EarlyStopping: 검증 손실이 5 에포크 동안 개선 없으면 자동 중단
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=50,           # 최대 50번 반복 학습
    batch_size=32,       # 한 번에 32개 샘플씩 처리
    validation_split=0.1, # 훈련 데이터의 10%를 검증용으로 사용
    callbacks=[early_stop],
    verbose=1
)

print('\n모델 학습 완료!')
```

```python
# 학습 과정 시각화 (손실 곡선)
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='훈련 손실')
plt.plot(history.history['val_loss'], label='검증 손실')
plt.title('모델 학습 손실 변화')
plt.xlabel('에포크')
plt.ylabel('손실(MSE)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# ─────────────────────────────────────────
# 8단계: 예측 및 평가
# ─────────────────────────────────────────

# 테스트 데이터로 예측
predicted_scaled = model.predict(X_test)

# 정규화를 원래 주가 단위로 되돌리기
predicted_price = scaler.inverse_transform(predicted_scaled)
actual_price = scaler.inverse_transform(y_test.reshape(-1, 1))

# 성능 평가 지표 계산
rmse = np.sqrt(mean_squared_error(actual_price, predicted_price))  # 평균 제곱근 오차
mae = mean_absolute_error(actual_price, predicted_price)            # 평균 절대 오차

print('=== 모델 성능 평가 ===')
print(f'RMSE (평균 예측 오차): {rmse:,.0f}원')
print(f'MAE  (평균 절대 오차): {mae:,.0f}원')
print(f'평균 주가 대비 오차율: {mae/actual_price.mean()*100:.1f}%')
```

```python
# 실제 주가 vs 예측 주가 시각화
test_dates = df.index[train_size + LOOKBACK:]

plt.figure(figsize=(14, 6))
plt.plot(test_dates, actual_price, label='실제 주가', color='steelblue', linewidth=1.5)
plt.plot(test_dates, predicted_price, label='예측 주가', color='red', linewidth=1.5, linestyle='--')
plt.title('삼성전자 주가: 실제 vs LSTM 예측', fontsize=14)
plt.xlabel('날짜')
plt.ylabel('주가 (원)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('\n분석 완료!')
```
